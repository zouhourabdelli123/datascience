```python
#bch nara kifh nsala7 les valeurs manquante mta3 Product_Category_2  w Product_Category_3
#bch nbadal les valeur null fel Product_Category_2 null mediande de Product_Category_2 bch matet2atharch bil outliers
#lezem na3ref moyenne wmediane 9add 9add ma8ir man7sb
#bch nsawar histogramme
#awal 7aga bch nimchi norsom il hit avec diff bins
#ken par rapport center copier coller rahou moyenne wmediane 9add 9add  sadpond mil bins hist maya3tich resultat s7i7aa selon nb d'observation selon pas mouch dima twassal
#bch n5dm awal 7aga 3al Product_Category_2
import matplotlib.pyplot as plt
for i in range (3,11):
    data["Product_Category_2"].hist(bins=i)
    plt.title("Product_Category_2 avec bins "+str(i))
    plt.show()
```

```python
#presque mouch 100% presque copier coller
# tawa bch nara famechy des outliers
data.boxplot(column="Product_Category_2")
```

```python
#ma3annech des outliers
#nb de lignie fel basa > 50 donc nsta3mal kstest
from scipy.stats import kstest
kstest(data['Product_Category_2'].dropna(),'norm')
#na7it les valeur null bil dropna snn ya3tini err
#pvalue=0.0 <0.05 => ya3ni mouch gaussiere so il moyenne mouch hiya mediane 
```

```python
#donc chnbaddal bil mediane
data['Product_Category_2'] = data['Product_Category_2'].fillna(data['Product_Category_2'].median())
```
