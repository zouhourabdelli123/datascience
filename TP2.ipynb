```python
#il pb hna ma3annech des histogramme copier coller 100%  rapport center
#heka alech n9oul on ne peut pas conclure wbch nmchi n'appliki test chapio-will bch y5aragli 2 valeur w7da statistique wa7da p-value ili hiya pb  ili il age variable gauussier
# ken il valeur <0.05 rahi mahiyech gaussier snn rahi mouch 100% s7i7a ama tgim tkoun
from scipy.stats import kstest
#bch n9ollou i7sb ma8ir matara null
kstest(data['Item_Weight'].dropna(),'norm')
#jena 0<0.05 donc on peut dire illi hiya impossible tkoun gaussier 
#ya3ni il moyenne wel mediane kifkif
#resultat bch n3awa4 il valeur manquante par la moyenne
```
