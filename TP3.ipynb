```python
from scipy import stats

data = [13, 12, 11, 12, 10.6, 15.75, 11, 14, 10, 11.25, 18, 13.5, 20, 15, 16.2]

ks_stat, ks_p = stats.kstest(data, 'norm')
print("Kolmogorov-Smirnov test:")
print(f"Statistique = {ks_stat}, p-value = {ks_p}")

shapiro_stat, shapiro_p = stats.shapiro(data)
print("\nShapiro-Wilk test:")
print(f"Statistique = {shapiro_stat}, p-value = {shapiro_p}")

normaltest_stat, normaltest_p = stats.normaltest(data)
print("\nD'Agostino et Pearson test:")
print(f"Statistique = {normaltest_stat}, p-value = {normaltest_p}")
```
