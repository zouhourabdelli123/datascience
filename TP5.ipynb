```python
#taw bch nrakab alih histg mta3 normale
#bch nthabat courbe mte3i s7i7a walla 
#normalment ta7t il courbe il adiyya
plt.hist(y, bins=3, density=True)
#density zdnehaa 5ater courb surf mte3ha 1 echelle taya7 par rapport azra9 
#density ya3ni zro99 il koll surf égalle 1 
#sna3t les pts
val_x = np.linspace(min(y), max(y), 100)

#koll pt sna3tha fel x isna3li img mte3ha fel y
from scipy.stats import norm
val_y = norm.pdf(val_x, np.mean(y), np.std(y))

plt.plot(val_x, val_y)

#sawart starr ydoll 3al moyenne
plt.axvline(np.mean(y), color='black', linestyle='dashed', linewidth=1)

plt.show()
```
